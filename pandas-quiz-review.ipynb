```python
import pandas as pd

df = pd.read_csv('https://gist.githubusercontent.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6/raw/92200bc0a673d5ce2110aaad4544ed6c4010f687/pokemon.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.HP.max()
```

```python
df[df.HP == df.HP.max()]
```

```python
def highest_attk(df):
    return df[df.HP == df.HP.max()]
highest_attk(df)
```

```python
df['Type 1'].nunique()
```

```python
def num_types(df):
    return df['Type 1'].nunique()
num_types(df)
```

```python
df['Type 2'].unique()
```

```python
ghosts = df[df['Type 2'] == 'Ghost']
len(ghosts)
```

```python
round((len(ghosts) / len(df))*100,2)
```

```python
def pct_ghost(df):
    gh = df[df['Type 2'] == 'Ghost']
    return round((len(gh) / len(df)*100),2)

pct_ghost(df)
```

```python
df['attck_diff'] = df.Attack - df.Defense
df.head()
```

```python
attackrs = df[df['attck_diff'] > 0]
attackrs.info()
```

```python
grss_n_rck = df[(df['type_1'] == 'Grass') | (df['type_1'] == 'Rock') | (df['type_2'] == 'Grass') |
                (df['type_2'] == 'Rock')]
grss_n_rck.sort_values('speed').nsmallest(1,'speed',keep='all')
```

```python
df = df.rename(columns = {        
    
                          '#': 'id',
                          'Name': 'name',
                          'Type 1': 'type_1',
                          'Type 2': 'type_2',
                          'Total': 'total',
                          'HP': 'hp',
                          'Attack': 'attack',
                          'Defense': 'defense',
                          'Sp. Atk': 'sp_atk',
                          'Sp. Def': 'sp_def',
                          'Speed': 'speed',
                          'Generation': 'generation',
                          'Legendary': 'legendary'
                                 }
                      )

df.head()
```

```python
df = df.rename(columns = {        
    
                          'sp_atk': 'special-attack',
                          'sp_def': 'special-defense'
                          }
)

df.head()
```

```python
df[df.attck_diff.abs() == df.attck_diff.abs().max()]
```

```python
df['sp_attck_diff'] = df['special-attack'] - df['special-defense']
df.sp_attck_diff.describe()
```

```python
df[df.sp_attck_diff.abs() == df.sp_attck_diff.abs().max()]
```

```python
t1rck_n_t2fry = df[(df['type_2'] == 'Fairy') & (df['type_1'] == 'Rock')]
t1rck_n_t2fry.info()
```
