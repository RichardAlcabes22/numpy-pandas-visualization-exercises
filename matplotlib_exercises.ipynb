```python
from matplotlib import pyplot as plt
import pandas as pd
from pydataset import data

## Data generation
import numpy as np
import random

## Mathematical functions
import math

## Random seed value to reproduce results
np.random.seed(123)
```

```python
# 1 plot the following : y = x2 - x + 2
x = list(range(0,20))
y = [((n ** 2)-(n)+2) for n in x]
print(x)
print(y)
```

```python
plt.plot(x,y)
plt.annotate('Origin (0,0)', xy =(0.2,8),xytext=(2,60),arrowprops={'facecolor':'black',
                                                                    'arrowstyle':'simple'})
plt.show();
```

```python
# 2 Create 4 charts with following equations
# y = x ** 0.5  y = x ** 3 y = log(base2)(x) y = 1/(1+x)

x1 = list(range(1,21))
y1 = [n ** 0.5 for n in x1]
print(x1)
print(y1)
```

```python
plt.scatter(x1,y1)
plt.show();
```

```python
x1 = list(range(1,21))
y2 = [n ** 3 for n in x1]
print(x1)
print(y2)
```

```python
plt.plot(x1,y2)
plt.show();
```

```python
x1 = list(range(1,21))
y3 = [math.log2(n) for n in x1]
print(x1)
print(y3)
```

```python
plt.scatter(x1,y3)
plt.show();
```

```python
x1 = list(range(1,21))
y4 = [1/(n+1) for n in x1]
print(x1)
print(y4)
```

```python
plt.scatter(x1,y4)
plt.show();
```

```python
# 3 combine 4 figures into one large figure with 4 subplots

n_rows = 2
n_cols = 2

plt.subplot(n_rows,n_cols,1)
plt.tight_layout()
plt.plot(x1,y1)
plt.title('$\sqrt{x}$')

plt.subplot(n_rows,n_cols,2)
plt.tight_layout()
plt.plot(x1,y2)
plt.title('$x^3$')

plt.subplot(n_rows,n_cols,3)
plt.tight_layout()
plt.plot(x1,y3)
plt.title('$\log_{2}$')

plt.subplot(n_rows,n_cols,4)
plt.tight_layout()
plt.plot(x1,y4)
plt.title('$1/(x+1)$')

plt.show();
```

```python
xx1 = list(np.arange(1,3,0.1))
yy2 = [round(n ** 3,2) for n in xx1]
```

```python
print(xx1)
print(yy2)
```

```python
# 4 Combine 4 figures with different color points all inot one graph, inlcude a proper legend and title
plt.scatter(x1,y1,alpha=0.4,edgecolors='black',label='$\sqrt{x}$')
plt.scatter(xx1,yy2,label='$x^3$')
plt.scatter(x1,y3,alpha=0.4,edgecolors='black',label='$\log_{2}$')
plt.scatter(x1,y4,label='$1/(x+1)$')
plt.ylim(-1,8)
plt.title('Various Options for UnSkewing Distributions')
plt.legend(loc=1)


plt.show;
```
