```python
# 5 Join employees and titles DFs together
```

```python
query_emp = 'SELECT * FROM employees'
employees_df = pd.read_sql(query_emp,get_db_url(env.uname,env.pwd,env.host,'employees'))
employees_df
```

```python
query_titles = 'SELECT * FROM titles'
titles_df = pd.read_sql(query_titles,get_db_url(env.uname,env.pwd,env.host,'employees'))
titles_df
```

```python
emp_titles_joind = pd.merge(employees_df, titles_df,how='inner',left_on='emp_no',right_on='emp_no')
emp_titles_joind
```

```python
# 6 find hire date of employee that is the most recent hire for each title

pd.pivot_table(data=emp_titles_joind,index='title',values='hire_date',aggfunc='max')
#.sort_values(
    #'item_price',ascending=False).nlargest(columns='item_price',n=1,keep='all')
```

```python
# 7 create a cross tab of number of titles by department
query_emp = 'SELECT * FROM dept_emp'
dpt_emp_df = pd.read_sql(query_emp,get_db_url(env.uname,env.pwd,env.host,'employees'))
dpt_emp_df
```

```python
query_emp = 'SELECT * FROM departments'
dpt_df = pd.read_sql(query_emp,get_db_url(env.uname,env.pwd,env.host,'employees'))
dpt_df
```

```python
emp_titles_dept_joind = pd.merge(emp_titles_joind, dpt_emp_df,how='inner',left_on='emp_no',right_on='emp_no')
emp_titles_dept_joind
```

```python
emp_titles_dept_joind_dept = pd.merge(emp_titles_dept_joind, dpt_df,how='inner',left_on='dept_no',right_on='dept_no')
emp_titles_dept_joind_dept
```

```python
pd.crosstab(emp_titles_dept_joind_dept.dept_name,emp_titles_dept_joind_dept.title)
```
