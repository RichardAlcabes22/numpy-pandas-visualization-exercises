```python
from pydataset import data
import numpy as np
import pandas as pd
```

```python
# 1 Copy the code from leasson to create dataframe of grades
students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas', 'Marie', 'Albert', 'Richard',
            'Isaac', 'Alan']
math_grades = np.random.randint(low=60, high=100, size=len(students))

english_grades = np.random.randint(low=60, high=100, size=len(students))

reading_grades = np.random.randint(low=60, high=100, size=len(students))

df_dict = {'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades}
df = pd.DataFrame(df_dict)
df
```

```python
# 1a create passing_english column
df['passing_english'] = df.english >= 70
df
```

```python
# 1b sort english grades by passing_english column how dups handled?
df.sort_values(by = 'passing_english')
# dups are handled by sorting on index (ascending)
```

```python
# 1c sort by mult criteria
df.sort_values(by = ['passing_english','name'])
```

```python
# 1d sort on passing_e then by grade
df.sort_values(by = ['passing_english','english'])
```

```python
# 1e calc Overall grade (mean) and add to df
df['Overall_Grade'] = round((df.english + df.reading + df.math) / 3, 1)
df
```

```python
# overall percent of specials

mamms_count = len(mammls_df.specials)
mamms_special_true_count = len(mammls_df[mammls_df.specials == True])

pct_special = round(mamms_special_true_count / mamms_count,2) * 100
```

```python
print(f'Overall Percent of Special Mammals is {pct_special}%')
```

```python
median_spd = mammls_df.describe().speed[mammls_df.describe().speed == 48.0]
median_spd
```

```python
fast_hoppers = len(mammls_df[(mammls_df.speed > mammls_df.speed.median()) & (mammls_df.hoppers == True)])
fast_hoppers
```

```python
fast_hoppers_pct = round(fast_hoppers / mamms_count,3) * 100
fast_hoppers_pct
```

```python
print(f'Overall Percent of Fast Hopping Mammals is {fast_hoppers_pct}%')
```
