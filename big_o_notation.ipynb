```python
from matplotlib import pyplot as plt
import pandas as pd
from pydataset import data

## Data generation
import numpy as np
import random

## Mathematical functions
import math

## Random seed value to reproduce results
np.random.seed(123)
```

```python
# 5 Graph the following: 
'''
y1 = On + 1  O(1)
y2 = log(n) O(log n)
y3 = n       O(log n)
y4 = n * log(n)   O(n log n)
y5 = n**2    O(n^2)
y6 = log2(n)  O(2^n)
y7 = n! O(n!)
y8 = n**n   O(n^n)
 
'''
```

```python
x1 = list(range(1,11))
y1 = [(n*0) + 1 for n in x1]
print(x1)
print(y1)
```

```python
x1 = list(range(1,11))
y2 = [math.log(n) for n in x1]
print(x1)
print(y2)
```

```python
x1 = list(range(1,11))
y3 = [n for n in x1]
print(x1)
print(y3)
```

```python
x1 = list(range(1,11))
y4 = [n * (math.log(n)) for n in x1]
print(x1)
print(y4)
```

```python
x1 = list(range(1,11))
y5 = [n**2 for n in x1]
print(x1)
print(y5)
```

```python
x1 = list(range(1,11))
y6 = [math.log(n,2) for n in x1]
print(x1)
print(y6)
```

```python
x1 = list(range(1,11))
y7 = [math.factorial(n) for n in x1]
print(x1)
print(y7)
```

```python
x1 = list(range(1,11))
y8 = [n**n for n in x1]
print(x1)
print(y8)
```

```python
plt.plot(x1,y1,label='O$(1)$')
plt.plot(x1,y2,label='O($\log n$)')
plt.plot(x1,y3,label='O$(n)$')
plt.plot(x1,y4,label='O($n (\log n)$)')
plt.plot(x1,y5,label='O$(n^2)$')
plt.plot(x1,y6,label='O$(\log_{2})$')
plt.plot(x1,y7,label='O$(n!)$')
plt.plot(x1,y8,label='O$(n^n)$')

plt.xlim(0,10)
plt.ylim(-10,100)
plt.title('BIG O NOTATION')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(loc=2)


plt.show;
```
