```python
swiss_df['is_catholic'] = swiss_df.Catholic >= 70
swiss_df.head()
```

```python
# interaction of catholic and fertility?
sns.catplot(
data=swiss_df,
x='is_catholic',
y='Fertility',
kind='box')
plt.show();
```

```python
# what measure correlates strongly with fertility?  Education and Examination status have the greatest absolute corr
# with fertility, in this case both are negatively correlated.  Strongest POS corr is Catholic.
swiss_raw = swiss_df.drop(columns='is_catholic')
sns.heatmap(swiss_raw.corr(),
           annot=True,
           cmap='coolwarm',
           vmax=1,
           vmin=-1,
           center=0,
           linewidth=1,
           linecolor='k',
           fmt='.2f')
plt.show()
```

```python
import env
query = 'SELECT * FROM orders'
```

```python
def get_db_url(u_name,pwd,h_name,db_name):
    url = f'mysql+pymysql://{u_name}:{pwd}@{h_name}/{db_name}'
    return url
```

```python
#next time use env.get_db_url!!!! duh... so thjat the function does not have to be defined inline
chip_df = pd.read_sql(query,get_db_url(env.uname,env.pwd,env.host,'chipotle'))
chip_df.head()
```

```python
chip_df['item_price'] = chip_df['item_price'].str.replace('$','').astype('float')
chip_df.info()
```

```python
chip_df['item_price'].head()
```

```python
top_4_items = pd.pivot_table(data=chip_df,index='item_name',values=['quantity','item_price'],aggfunc='sum').sort_values(
    'quantity',ascending=False).nlargest(columns='quantity',n=4,keep='all')
top_4_items
top_4_items.reset_index(inplace=True)
```

```python
top_4_items
```

```python
sns.catplot(
data=top_4_items,
x='item_name',
y='item_price',
kind='bar')

plt.show();
```
